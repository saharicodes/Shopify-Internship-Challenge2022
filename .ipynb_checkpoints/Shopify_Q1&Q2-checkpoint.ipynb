**Question 1**: Given some sample data, write a program to answer the following:

On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis. 

What metric would you report for this dataset?
What is its value?


**a.** Think about what could be going wrong with our calculation. Think about a better way to evaluate this data. 

**Answer:** After exploring the data set, it turned out that the calculated AOV of $3145.13 is the average of `order_amount`. Since orders may have more than one item i.e.`total_items`, the calculated value is not the correct estimate of the price of one shoe.

**b.** What metric would you report for this dataset? 

**Answer:** Since each shop only sells one model of a shoe, we can calculate the price of one shoe per order by deviding `order_amount` by `total_items` and then calculate the average price of a shoe across orders as a metric.


**c.** What is its value?

**Answer:** To calculate the average price of a shoe, we have to examine data first to find outliers. Since outliers have negative effect on the average shoe price, they need to be removed from data set. The calculation process is as the following: 

````python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# magic command for matplotlib in notebooks
%matplotlib inline

# Load data set 
data = pd.read_csv("2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv")

# Display data set
display(data.head(n=5))

# Determine if there is any null values
print("Number of missing values: {}".format(data.isnull().values.sum()))
````

````python
# Total number of orders
n_orders = data.shape[0]
print(f"Total number of orders {n_orders}")

# Price of one shoe in each order
data['shoe_price']= data['order_amount']/data['total_items']
display(data.head(n=5))
````

````python
# Visualize distribution of shoe price
plt.hist(data['shoe_price'],bins = 30, color = '#00A0A0')
plt.xlabel('Shoe price per order')
plt.ylabel('Number of orders')
plt.title('Shoe price distribution')
plt.show()
````

````python
# Perform box plot and identify outliers
plt.boxplot(data['shoe_price'])
plt.show

# calculate summary statistics
data_mean, data_std = data['shoe_price'].mean(), data['shoe_price'].std()
# identify outliers
cut_off = data_std * 3
lower, upper = data_mean - cut_off, data_mean + cut_off
# identify outliers
outliers = [x for x in data['shoe_price'] if x < lower or x > upper]
print(f"Number of outliers {len(outliers)}")

# remove outliers
outliers_removed = [x for x in data['shoe_price'] if x >= lower and x <= upper]
print(f"Number of non-outliers {len(outliers_removed)}")
````

````python
plt.boxplot(pd.Series(outliers_removed))
plt.show
````

````python
# Calculate metric that is average shoe price per order 

print(f"Average shoe price per order when data has outliers: {data['shoe_price'].mean()}")
print(f"Average shoe price per order when outliers removed: {pd.Series(outliers_removed).mean()}")
````

**c.** What product was ordered the most by customers in Germany?

Querry

```
SELECT ProductID, Country, MAX(product_count)
```
